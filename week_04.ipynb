### Frequency Distribution
- 단어의 빈도수를 활용할 수 있는 많은 메소드 제공
- 인간언어 -> 컴퓨터 인식
    - 전체 단어 수
    - 특정 단어 빈도수

```python
fdist = FreqDist(text1)
fdist # dictionary = key:value
```

```python
fdist.most_common(10) #가장 많이나오는 단어
```

```python
fdist['whale'], text1.count('whale')
```

```python
fdist.freq('whale') #단어 출현비율
```

```python
fdist.plot(50, cumulative=True) #(제일 빈번한 단어 1~50까지, 축적(False로하면 그냥 출현횟수))
```

```python
# 1번 나타나는 단어들
# fdist.hapaxes() 
len(fdist.hapaxes()) #중복 안됨, 대소문자 구분됨
```

### list_comprehension
- 지능형 리스트
- 주어진 리스트를 포함하는 새로운 리스트를 형성하는 것
- list + for_loop 한줄로 해결하는 방식
- [A for A' in B] 
   - B에 있는 A'에서 A를 가진 리스트를 반환
   - B: 기존의 리스트
   - A': B의 자료 형태
   - A: 새로운 리스트의 요소
- 조건+ :[A for A' in B if<조건식>]
    - [A for A' in B if <조건식> and <조건식>,...]

```python
[d for d in drinks]
```

```python
[d[0] for d in drinks]
```

```python
# text1 알파벳 수가 10이상인 단어
[word for word in text1 if len(word) >= 10]
```

```python
# text4에서 'gnt'를 포함하는 단어 반환
[word for word in text4 if 'gnt' in word]
```

```python
# 중복 없앰
print([word for word in set(text4) if 'gnt' in word]) #중복없애고 조건실행, 리스트
print(set([word for word in text4 if 'gnt' in word])) #조건 실행후 중복없앰, 세트
```

### 출력 포매팅
- 1. f'str...{var}'
    -f, F 둘 다 가능
- 2. % # 옛날방식
    -'...%s' %var
- 3. '...{}'.format(var)

```python
name = 'lee'
age = 14
print(f'my name is {name}')
```

```python
print(f'my name is {name}, and I am {age}')
```

```python
print('my name is %s' %name)
```

```python
print('my name is {}'.format(name))
```

```python
print('my name is {1}, and I am {0}'.format(name, age))
```

```python
print('my name is {:.2}, and I am {}'.format(a, age))
```

```python
a = 0.555555
print('my name is %.3f' %a)
```
